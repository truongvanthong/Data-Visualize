Vì vậy, khung dữ liệu bao gồm 150.500 vụ phạm tội, diễn ra vào năm 2016. Để giảm chi phí tính toán, chúng ta chỉ cần làm việc với 100 sự cố đầu tiên trong tập dữ liệu này.


````python
# get the first 100 crimes in the df_incidents dataframe
limit = 100
df_incidents = df_incidents.iloc[0:limit, :]
````

Hãy khẳng định là tập dữ liệu của chúng ta chỉ chứa 100 vụ phạm tội.


````python
df_incidents.shape
````

Bây giờ chúng ta đã giảm bớt dữ liệu một chút, hãy hình dung nơi những tội ác này đã diễn ra ở thành phố San Francisco. Chúng ta sẽ sử dụng kiểu mặc định và chúng ta sẽ khởi tạo mức thu phóng thành 12.


````python
# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42
````

````python
# create map and display it
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# display the map of San Francisco
sanfran_map
````

Bây giờ chúng ta hãy chồng các vị trí của vụ phạm tội lên bản đồ. Cách để thực hiện điều đó trong **Folium** là tạo một _feature group_ với các đặc tính và style riêng, sau đó thêm nó vào sanfran_map.


````python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
sanfran_map.add_child(incidents)
````

Bạn cũng có thể thêm một số văn bản pop-up sẽ được hiển thị khi bạn di chuột qua một điểm đánh dấu. Hãy làm cho mỗi điểm đánh dấu hiển thị danh mục tội phạm khi di chuột qua.

````python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add pop-up text to each marker on the map
latitudes = list(df_incidents.Y)
longitudes = list(df_incidents.X)
labels = list(df_incidents.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map)    
    
# add incidents to map
sanfran_map.add_child(incidents)
````

Cũng không thú vị lắm có phải không? Bây giờ bạn có thể biết loại tội phạm nào xảy ra ở mỗi điểm đánh dấu.

Nếu bạn thấy bản đồ có quá nhiều điểm đánh dấu, thì có hai giải pháp cho vấn đề này. Giải pháp đơn giản hơn là xóa các điểm đánh dấu vị trí này và chỉ cần thêm văn bản vào chính các điểm đánh dấu vòng tròn như sau:


````python
# create map and display it
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# loop through the 100 crimes and add each to the map
for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5, # define how big you want the circle markers to be
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(sanfran_map)

# show map
sanfran_map
````

Cách khắc phục thích hợp khác là nhóm các điểm đánh dấu thành các cụm khác nhau. Mỗi cụm sau đó được đại diện bởi số lượng tội phạm trong mỗi khu phố. Các cụm này có thể được coi là các túi của San Francisco mà sau đó bạn có thể phân tích riêng.

Để thực hiện điều này, chúng ta bắt đầu bằng cách khởi tạo một đối tượng _MarkerCluster_ và thêm tất cả các điểm dữ liệu trong khung dữ liệu vào đối tượng này.


````python
from folium import plugins

# let's start again with a clean copy of the map of San Francisco
sanfran_map = folium.Map(location = [latitude, longitude], zoom_start = 12)

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(sanfran_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# display map
sanfran_map
````

Lưu ý khi bạn thu nhỏ hết mức, tất cả các điểm đánh dấu được nhóm thành một cụm, _the global cluster_, gồm 100 điểm đánh dấu hoặc tội phạm, là tổng số tội phạm trong khung dữ liệu của chúng ta. Để tạo một bản đồ `Choropleth`, chúng ta cần một file GeoJSON định nghĩa diện tích hoặc biên giới mỗi bang, quận hoặc quốc gia mà chúng ta quan tâm. Trong trường hợp này, vì chúng ta đang nỗ lực tạo bản đồ thế giới, chúng ta muốn một GeoJSON xác định ranh giới của tất cả các quốc gia trên thế giới. Để thuận tiện cho bạn, chúng tôi sẽ cung cấp cho bạn tệp này, vì vậy hãy tiếp tục và tải xuống. Hãy đặt tên cho nó là **world_countries.json**.



````python
# download countries geojson file
#!wget --quiet https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/world_countries.json
    
#print('GeoJSON file downloaded!')
````

Bây giờ, chúng ta đã có file GeoJSON, hãy tạo một bản đồ thế giới, lấy trung tâm là giá trị _vĩ tuyến_ và _kinh tuyến_ **[0, 0]**  với mức độ thu phóng ban đầu là 2.


````python
world_geo = r'world_countries.json' # geojson file

# create a plain world map
world_map = folium.Map(location=[0, 0], zoom_start=2)
````

Và để tạo một biểu đồ `Choropleth`, chúng ta sẽ sử dụng phương thức _choropleth_ với các tham số chính sau:

1.  geo_data, chính là file GeoJSON.
2.  data, chính là khung dữ liệu chứa dữ liệu.
3.  columns, biểu diễn các cột trong khung dữ liệu sẽ được sử dụng để tạo bản đồ `Choropleth`.
4.  key_on, chính là khóa hoặc biến trong file GeoJSON chứa tên biến ưa thích.Để xác định điều đó, bạn sẽ cần mở tệp GeoJSON bằng bất kỳ trình soạn thảo văn bản nào và ghi lại tên của khóa hoặc biến có chứa tên của các quốc gia, vì các quốc gia là biến mà chúng ta quan tâm. Trong trường hợp này, **name** là khóa trong tệp GeoJSON chứa tên của các quốc gia. Lưu ý rằng khóa này là phân biệt chữ hoa chữ thường, vì vậy bạn cần nhập vào chính xác như nó tồn tại trong tệp GeoJSON.


````python
# generate choropleth map using the total immigration of each country to Canada from 1980 to 2013
world_map.choropleth(
    geo_data=world_geo,
    data=df_can,
    columns=['Country', 'Total'],
    key_on='feature.properties.name',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Immigration to Canada'
)

# display map
world_map
````

Theo chú giải bản đồ `Choropleth` của chúng ta, màu của quốc gia càng tối và màu càng gần với màu đỏ thì số lượng người nhập cư từ quốc gia đó càng cao. Theo đó, lượng nhập cư cao nhất trong vòng 33 năm (từ 1980 đến 2013) là từ Trung Quốc, Ấn Độ và Philippines, tiếp theo là Ba Lan, Pakistan và thú vị là  có cả nước Mỹ.


Lưu ý cách chú giải đang hiển thị ranh giới hoặc ngưỡng âm. Hãy khắc phục điều đó bằng cách xác định các ngưỡng của riêng chúng ta và bắt đầu bằng 0 thay vì -6,918!


````python
world_geo = r'world_countries.json'

# create a numpy array of length 6 and has linear spacing from the minium total immigration to the maximum total immigration
threshold_scale = np.linspace(df_can['Total'].min(),
                              df_can['Total'].max(),
                              6, dtype=int)
threshold_scale = threshold_scale.tolist() # change the numpy array to a list
threshold_scale[-1] = threshold_scale[-1] + 1 # make sure that the last value of the list is greater than the maximum immigration

# let Folium determine the scale.
world_map = folium.Map(location=[0, 0], zoom_start=2)
world_map.choropleth(
    geo_data=world_geo,
    data=df_can,
    columns=['Country', 'Total'],
    key_on='feature.properties.name',
    threshold_scale=threshold_scale,
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Immigration to Canada',
    reset=True
)
world_map
````

Bây giờ đã tốt hơn nhiều rồi! Hãy thoải mái tìm hiểu dữ liệu và có thể tạo bản đồ `Choropleth` trong nhiều năm, hoặc có thể là nhiều thập kỷ, và hãy so sánh chúng với toàn bộ giai đoạn từ 1980 đến 2013.
